```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load CSV file (example)
df = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_76034.csv', skiprows=4)
df.head()
```

```python
data = df[['Country Name', '2022']].dropna()
data.columns = ['Country', 'Population']
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Population'], bins=30, kde=True)
plt.title('Population Distribution Across Countries (2022)', fontsize=14)
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.tight_layout()
plt.show()
```
